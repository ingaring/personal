```python
import numpy as np
import random as rand
from matplotlib.patches import RegularPolygon
import matplotlib.pyplot as plt

# Input
width = 20
height = 20

start_disperging = height/5
gradient = "linear"  #"2poly" or "linear" 

X = np.array(range(width))
Y = np.array(range(height))*0.9
```

```python

fig = plt.figure()
ax = fig.add_subplot(111)

for y in Y:   
    if y > start_disperging:
        if gradient == "linear": 
            grad = (y-start_disperging)/(height-start_disperging)
        if gradient == "2poly": 
            grad = ((y-start_disperging)/(height-start_disperging))**2

        for x in X:
            if (y/0.9)%2: x+=0.5
            randnr = rand.random()    
            if grad < randnr:
                color = "green"
                hex = RegularPolygon((x, y), numVertices=6, radius=0.52, 
                                     #orientation=np.radians(30), 
                                     facecolor=color, alpha=0.2, edgecolor='k')
                ax.add_patch(hex)
            else:

                color = "white"
                hex = RegularPolygon((x, y), numVertices=6, radius=0.52, 
                                     #orientation=np.radians(30), 
                                     facecolor=color, alpha=0.2, edgecolor='k')
                ax.add_patch(hex)
    else:
        for x in range(width):
            if (y/0.9)%2: x+=0.5
            color = "green"
            hex = RegularPolygon((x, y), numVertices=6, radius=0.52, 
                                     #orientation=np.radians(30), 
                                     facecolor=color, alpha=0.2, edgecolor='k')
            ax.add_patch(hex)

ax.set_xlim([-1,width])
ax.set_ylim([-1,height])
plt.show()
```
